# Problem 1: Multiples of 3 or 5
https://projecteuler.net/problem=1

Find the sum of all the multiples of 3 or 5 below 1000.

```python
# Find the sum of all the multiples of 3/5 up to the input
def multiples_of_3_or_5(x):
	return sum([i for i in range(x + 1) if i % 3 == 0 or i % 5 == 0])

# Problem 1: Find the sum of all the multiples of 3/5 below 1000.
print("Sum: " +str(multiples_of_3_or_5(1000)))
```

# Problem 2: Even Fibonacci numbers
https://projecteuler.net/problem=2

By considering the terms in the Fibonacci sequence whose values do not exceed four million, find the sum of the even-valued terms.

```python
# Recursively returns the fibonnaci value based on the input.
def fibonacci(x):
	if x <= 2:
		return x
	return fibonacci(x - 1) + fibonacci(x - 2)

sum = 0
fib_0, fib_1 = 0, 1 # Initial value for the fibonacci sequence
while sum <= 4000000:
	fib = fib_0 + fib_1
	# Only add to the sum if the resulting fibonacci value is even
	if fib % 2 == 0:
		sum += fib
	# Setting new values
	fib_0 = fib_1
	fib_1 = fib

print("Sum: " +str(sum))
```
